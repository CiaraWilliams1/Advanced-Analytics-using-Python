```python
# Import the necessary libraries, packages, and modules.
# Data analysis:
import numpy as np  
import pandas as pd 

# Linear regression:
import sklearn
import statsmodels.api as sm 
from sklearn.model_selection import train_test_split 

# Visualisation:
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Import the salary_data.csv file.
data = pd.read_csv('ecommerce_data.csv') 

# Print the DataFrame.
(data.head())
```

```python
data2 = data[['avg_no_it', 'tax', 'Median_s']]
data2.head()
```

```python
# Replace the missing values with 0.
data2.fillna(0, inplace=True)

# Determine the number of missing values.
data2.isna().sum()
```

## 3. Define the variables

```python
# Define the dependent variable.
y = data2['Median_s'].values.reshape(-1, 1) 

# Define the independent variable.
x = data2['avg_no_it'].values.reshape(-1, 1) 
```

## 4. Split the data set

```python
# Create  the subset (50/50); 
# Control the shuffling/avoid variation in values between variables.
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    train_size=0.7,
                                                    random_state=42)
```

## 5. Run a linear equation

```python
# Import the necessary package.
from sklearn.linear_model import LinearRegression

# Specify the model.
lr = LinearRegression()

# Fit the model.
lr.fit(x_train, y_train) 

# Predict the training set values.
y_pred = lr.predict(x_train) 

# View the output.
y_pred
```

## 6. Plot the regression

```python
# Create a scatterplot with a regression line.
plt.scatter(x_train, y_train, color = 'red')  
plt.plot(x_train, y_pred, color = 'green')

# Set the title and legends for the graph.  
plt.title("Median value of business by avg number of items sold")
plt.xlabel("Avg items sold")
plt.ylabel("Median value") 

# Print the graph. 
plt.show() 
```

```python
# Predict the training set values.
y_pred_test = lr.predict(x_test) 

# View the output.
y_pred_test
```

```python
# Create a scatterplot with a regression line.
plt.scatter(x_test, y_test, color = 'red')  
plt.plot(x_test, y_pred_test, color = 'green')

# Set the title and legends for the graph.  
plt.title("Median value of business by avg number of items sold")
plt.xlabel("Avg items sold")
plt.ylabel("Median value") 

# Print the graph. 
plt.show() 
```

```python
# Print R-squared value of the test data.
print("R-squared value: ", lr.score(x_test, y_test)) 
print("R-squared value: ", lr.score(x_train, y_train)) 

# Print the intercept value.
print("Intercept value: ", lr.intercept_) 

# Print the coefficient value.
print("Coefficient value: ", lr.coef_)
```

#### Notes:
- The R-squared tells us that the model is explaining 45.85% of the model.
- The coefficient value of 9.12 tells us that as the `lowstat` variable increases by 1, the predicted value of `Median_s` increases by 9.12.
